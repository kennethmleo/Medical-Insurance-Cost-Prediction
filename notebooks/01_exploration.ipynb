```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", palette="pastel")
```

```python
df = pd.read_csv("../data/medical_insurance.csv")
df.head()
```

```python
df.shape
df.info()
df.isnull().sum().sort_values(ascending=False).head(20)
```

```python
df.describe(include='all').transpose().head(20)
```

```python
sns.histplot(df['monthly_premium'], kde=True)
plt.title("Distribution of Monthly Premium")
plt.xlabel("Monthly Premium")
plt.show()

print(df['monthly_premium'].describe())
```

```python
corr = df.corr(numeric_only=True)['monthly_premium'].sort_values(ascending=False)
corr.head(15)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes('number').corr(), cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='smoker', y='monthly_premium', data=df)
plt.title("Monthly Premium by Smoker Status")

sns.boxplot(x='plan_type', y='monthly_premium', data=df)
plt.xticks(rotation=45)
plt.title("Monthly Premium by Plan Type")
plt.show()

sns.boxplot(x='region', y='monthly_premium', data=df)
plt.xticks(rotation=45)
plt.title("Monthly Premium by Region")
plt.show()
```

```python
sns.scatterplot(x='age', y='monthly_premium', data=df)
plt.title('Age vs Monthly Premium')

sns.scatterplot(x='bmi', y='monthly_premium', data=df)
plt.title('BMI vs Monthly Premium')

sns.scatterplot(x='risk_score', y='monthly_premium', data=df)
plt.title('Risk Score vs Monthly Premium')
plt.show()
```

```python
corr_matrix = df.corr(numeric_only=True)
high_corr = corr_matrix[(corr_matrix.abs() > 0.95) & (corr_matrix.abs() < 1.0)]
high_corr.dropna(how="all", inplace=True)
high_corr
```
